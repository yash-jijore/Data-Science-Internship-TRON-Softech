```python
import pandas as pd
```

```python
data = {
    "Student" : ['A' ,'B' , 'C' , 'D' , 'E' ] ,
    "Math" : [78,65,90,55,82] ,
    "Science" : [85,70,92,60,80] , 
    "English" : [74,68,88,58,79]
}

df = pd.DataFrame(data)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(df['Student'].astype('str'),df['Average'] )
plt.xlabel('Student')
plt.ylabel('Average Marks')
plt.title('Average Marks of Student')
plt.tight_layout()
plt.grid(True)
```

```python
plt.bar(sub_avg.index ,sub_avg.values )
plt.xlabel('Subject')
plt.ylabel('Average Marks')
plt.title('Subject Wise Average')
plt.tight_layout()
plt.grid(True)
```

```python
plt.figure()
plt.plot(df['Student'].astype('str'), df['Math'], label='Math')
plt.plot(df['Student'].astype('str'), df['Science'], label='Science')
plt.plot(df['Student'].astype('str'), df['English'], label='English')
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Marks Comparison")
plt.legend()
plt.grid(True)
```

```python
import pandas as pd 
import numpy as n
```

```python
data = {
    "Student" : ['A' ,'B' , 'C' , 'D' , 'E' ] ,
    "Math" : [78,65,90,55,82] ,
    "Science" : [85,70,92,60,80] , 
    "English" : [74,68,88,58,79]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
df['Average'] = df[['Math', 'Science', 'English']].mean(axis=1)
print(df)
```

```python
sub_avg = df[['Math' , 'Science' , 'English']].mean() 
sub_avg
#sub_avg.info()
```

```python
id = df['Average'].idxmax()
top_stud = df.iloc[id]
```

```python
top_stud
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x='Student' , y = 'Average' , data = df)
plt.title('Average marks of student')
plt.tight_layout()
plt.grid(True)
```

```python
plt.bar(sub_avg.index , sub_avg.values)
plt.xlabel('Subejcts')
plt.ylabel('Marks')
plt.title('Average Marks Per Subject')
plt.tight_layout()
plt.grid(True)
```

```python
print('Top Student is : ' , top_stud['Student'])
print('Highest Average Marks is in : ' , sub_avg.idxmax())
```

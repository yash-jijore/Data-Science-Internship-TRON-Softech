```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://truerp.in/index.php'
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/120.0 Safari/537.36"
}

response = requests.get(url , headers = headers)
response.status_code
```

```python
soup = BeautifulSoup(response.text , 'html.parser')
```

```python
title = soup.select_one('#hero h1').text.strip()
desc = soup.select_one("#hero p").text.strip()
```

```python
df = pd.DataFrame([{
    "Title" : title , 
    "Description" : desc
}])
```

```python
df.to_csv("Hero.csv" , index=False)
```

```python
about = soup.select_one("#about").text.strip()

df1 = pd.DataFrame([{
    "About": about
}])
df1.to_csv("About.csv", index=False)
```

```python
features = soup.select("#features .feature-item")
feature_list = []

for f in features:
    feature_list.append(f.text.strip())

features_df = pd.DataFrame(feature_list, columns=["Feature"])
features_df.to_csv("Features.csv", index=False)
```

```python
services = soup.select("#services .service-item")
service_list = []

for s in services:
    service_list.append(s.text.strip())

services_df = pd.DataFrame(service_list, columns=["Service"])
services_df.to_csv("Services.csv", index=False)
```

```python
stats = soup.select("#stats .stats-item")
stat_list = []

for stat in stats:
    value = stat.select_one("span").text.strip()
    label = stat.select_one("p").text.strip()
    stat_list.append([label, value])

stats_df = pd.DataFrame(stat_list, columns=["Metric", "Value"])
stats_df.to_csv("Stats.csv", index=False)
```

```python
faqs = soup.select("#faq .faq-item")
faq_list = []

for faq in faqs:
    question = faq.select_one("h3").text.strip()
    answer = faq.select_one("p").text.strip()
    faq_list.append([question, answer])

faq_df = pd.DataFrame(faq_list, columns=["Question", "Answer"])
faq_df.to_csv("Faq.csv", index=False)
```

```python
import pandas as pd
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_excel('iris-train.xlsx')
```

```python
df.head()
```

```python
df.tail()
```

```python
x = df.iloc[:,:-1]
```

```python
x
```

```python
y=df.iloc[:,-1]
```

```python
y
```

```python
model = RandomForestClassifier()
```

```python
model = model.fit(x.values ,y.values)
```

```python
model
```

```python
predictions = model.predict([[6	,3.4	,4.5	,1.6]])
```

```python
predictions
```

```python
df_test = pd.read_excel('iris-test.xlsx')
```

```python
df_test_1 = df_test.iloc[:,:-1]
df_test_1
```

```python
predictions = model.predict(df_test_1.values)
```

```python
predictions
```

```python
y_test = df_test.iloc[:,-1]
```

```python
actual = y_test.values
```

```python
count = 0
for i in range(0 , len(actual)):
    if actual[i] == predictions[i]:
        count = count+1
print(f"The Accuracy of the Model is : {(count*100)/len(actual)}")        
```
